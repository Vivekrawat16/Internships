```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
# print(df.head())
print(df.info())
# print(df.describe())
```

```python
sns.barplot(x='category', y='Total', data=df, estimator=sum)
plt.title("Total Sales by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
df['day_of_week'] = df['invoice_date'].dt.day_name()

# Sales by day
sns.barplot(data=df, x='day_of_week', y='Total', estimator=sum, order=[
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title("Sales by Day of Week")
plt.xticks(rotation=45)
plt.show()
```

```python
df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True, errors='coerce')
```

```python
df['month'] = df['invoice_date'].dt.month
df['year'] = df['invoice_date'].dt.year

# Monthly sales
monthly_sales = df.groupby(['year', 'month'])['Total'].sum().reset_index()

sns.lineplot(data=monthly_sales, x='month', y='Total', hue='year')
plt.title("Monthly Sales Over Years")
plt.show()
```

```python
bins = [0, 18, 25, 35, 45, 60, 100]
labels = ['<18', '18-25', '26-35', '36-45', '46-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Age group spending
sns.barplot(data=df, x='age_group', y='Total', estimator=sum)
plt.title("Total Spend by Age Group")
plt.show()
```

```python
sns.barplot(data=df, x='gender', y='Total', estimator=sum)
plt.title("Total Spend by Gender")
plt.show()
```

```python
sns.countplot(data=df, x='payment_method', order=df['payment_method'].value_counts().index)
plt.title("Preferred Payment Methods")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(data=df, x='category', y='Total', estimator=sum)
plt.title("Sales by Product Category")
plt.xticks(rotation=45)
plt.show()
```

```python
category_avg = df.groupby('category')['Total'].mean().sort_values(ascending=False)

sns.barplot(x=category_avg.index, y=category_avg.values)
plt.title("Average Spend per Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# Select numeric columns
num_cols = ['quantity', 'price', 'Total', 'age']

# Correlation matrix
corr = df[num_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python


# Aggregate daily sales
daily_sales = df.groupby('invoice_date')['Total'].sum()

# Ensure proper datetime index
daily_sales.index = pd.to_datetime(daily_sales.index)

# STL Decomposition
stl = STL(daily_sales, period=30)  # approx. 1-month seasonality
result = stl.fit()

# Plot the decomposition
result.plot()
plt.suptitle("STL Decomposition of Daily Sales")
plt.show()
```
