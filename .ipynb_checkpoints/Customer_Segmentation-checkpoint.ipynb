```python
import pandas as pd
from datetime import timedelta
df = pd.read_csv('cleaned_data.csv')
```

```python
today = df['invoice_date'].max()
df['invoice_date'] = pd.to_datetime(df['invoice_date'])
today = df['invoice_date'].max()  
```

```python
rfm = df.groupby('customer_id').agg({
    'invoice_date': lambda x: (today - x.max()).days,      # Recency
    'invoice_no': 'nunique',                                # Frequency
    'Total': 'sum'                                          # Monetary
}).reset_index()

rfm.columns = ['customer_id', 'recency', 'frequency', 'monetary']
```

```python
# Normalize frequency to 0–100 scale
rfm['loyalty_score'] = (rfm['frequency'] - rfm['frequency'].min()) / (rfm['frequency'].max() - rfm['frequency'].min()) * 100

# Step 1: Get most used payment method per customer
payment_pref = df.groupby(['customer_id', 'payment_method']).size().reset_index(name='count')
payment_pref = payment_pref.loc[payment_pref.groupby('customer_id')['count'].idxmax()]
payment_pref = payment_pref[['customer_id', 'payment_method']]

# Step 2: Drop 'payment_method' from rfm if it already exists
if 'payment_method' in rfm.columns:
    rfm = rfm.drop(columns=['payment_method'])

# Step 3: Merge the most used payment method into rfm
rfm = pd.merge(rfm, payment_pref, on='customer_id', how='left')

# Step 4: Encode payment method
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
rfm['payment_encoded'] = le.fit_transform(rfm['payment_method'])
```

```python
from sklearn.preprocessing import StandardScaler

# Step 1: Select features
# Note: Ensure 'avg_discount_rate' column exists in rfm — or remove it if not used
features = rfm[['recency', 'frequency', 'monetary', 'loyalty_score', 'payment_encoded']].fillna(0)

# Step 2: Normalize the selected features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=4, random_state=42)
rfm['gmm_segment'] = gmm.fit_predict(scaled_features)
```

```python
# Group by cluster and summarize
segment_summary = rfm.groupby('gmm_segment').agg({
    'recency': 'mean',
    'frequency': 'mean',
    'monetary': 'mean',
    'loyalty_score': 'mean',
    
}).round(2)

print(segment_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='gmm_segment', y='monetary', data=rfm)
plt.title("Monetary Value by GMM Segment")
plt.show()

sns.countplot(x='gmm_segment', hue='payment_method', data=rfm)
plt.title("Payment Preference by Segment")
plt.show()
```

```python
# Recalculate loyalty score
rfm['loyalty_score'] = (rfm['frequency'] - rfm['frequency'].min()) / (rfm['frequency'].max() - rfm['frequency'].min()) * 100

# Recalculate segment summary
segment_summary = rfm.groupby('gmm_segment').agg({
    'recency': 'mean',
    'frequency': 'mean',
    'monetary': 'mean',
    'loyalty_score': 'mean'
}).round(2)

print(segment_summary)
```
